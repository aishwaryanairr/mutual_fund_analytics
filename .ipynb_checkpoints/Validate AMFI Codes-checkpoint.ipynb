```python
import pandas as pd
```

```python
fund_master = pd.read_csv(
    "data/raw/01_fund_master.csv"
)

nav_history = pd.read_csv(
    "data/raw/02_nav_history.csv"
)
```

```python
fund_codes = set(
    fund_master["amfi_code"]
)

nav_codes = set(
    nav_history["amfi_code"]
)

missing_codes = fund_codes - nav_codes

print("Missing Codes")
print(missing_codes)
```
